```python
import csv

f = open("card.csv", encoding='UTF8')
data = csv.reader(f)
next(data)
data=list(data) # 읽어들일 데이터 리스트로 변환

print("#1번")
print(type(data))

print("#2번")
print(len(data))

print("#3번")
print(data[0])

print("#4번")
print(data[:3])

print("#5번")
for line in data:
  print(line[0],line[5],line[6])

print("#6번")
sum=0
for row in data:
  date = row[0]
  month = date.split('-')[1]
  if(month == '12'):
    sum+=int(row[6]) #파일에서 읽은 값은 문자열로 인식하기 때문
print("2019년년 12월 총 이용 금액: ",sum,"원")

f.close()
```
