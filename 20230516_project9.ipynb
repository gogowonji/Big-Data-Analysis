시도구군만 가져오기

```python
addr_sigu = addr.drop(['그외주소'],axis=1)
addr_sigu.head()
```

```python
addr_sigu['count']=0
addr_group=addr_sigu.groupby(['시도','구군','시도구군'], as_index=False).count()
addr_group.head()
```

```python
# 시도구군별로 의료기관 알고 싶기 때문에 인덱스 변경
addr_group=addr_group.set_index('시도구군')
addr_group.head()
```

행정구역별 인구수 가져오기

```python
population=pd.read_excel('행정구역_시군구_별__성별_인구수.xlsx')
population.head()
```

```python
population=population.rename(columns={'행정구역(시군구)별':'시도','행정구역(시군구)별.1':'구군'})
#데이터 병합을 위해 컬럼이름 동일하게 재작성
population.head()
```

```python
for element in range(0,len(population)):
  population['시도'][element]=population['시도'][element].strip()
  population['구군'][element]=population['구군'][element].strip()
#시도, 군구 열 값들이 오른쪽 정렬, 왼쪽 공백 없애는 작업
population['시도구군']=population.apply(lambda r:r['시도']+' '+r['구군'],axis=1)
#시도+군구=시도군구 열 생성
population=population[population.구군 != '소계'] #구군 열 값이 '소계'인 것은 제외
population=population.set_index('시도구군')
population.head()
```

```python
addr_population_merge=pd.merge(addr_group, population, left_index = True, right_index = True)
#시도구군이 있는 데이터들만 합쳐줘
addr_population_merge.head()
#동일한 컬럼이 있다면 _x, _y가 붙음
```

```python
local_population = addr_population_merge[['시도_x','구군_x','count','총인구수 (명)']]
local_population.head()
```

```python
local_population = local_population.rename(columns = {'시도_x':'시도','구군_x':'구군','총인구수 (명)':'인구수'})
count = local_population['count']
local_population['ratio'] = count.div(local_population['인구수'], axis=0)*10000
local_population.head()
```

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams , style
style.use('ggplot') # 밑에 스타일 종류 참고
from matplotlib import font_manager , rc
font_name = font_manager.FontProperties(fname ="malgun.ttf").get_name()
rc('font', family = font_name)
```

```python
MC_count= local_population [['count']]
MC_count = MC_count.sort_values ('count', ascending = False) #'count' 칼럼을 기준으로 내림차순
plt.rcParams['figure.figsize']=(25,5)
MC_count.plot(kind='bar',rot =90)
plt.show
```

```python
MC_ratio = local_population[['ratio']]
MC_ratio = MC_ratio.sort_values ('ratio', ascending = False) #ratio' 칼럼을 기준으로 내림차순
plt.rcParams['figure.figsize']=(25,5)
MC_ratio.plot(kind='bar',rot =90)
plt.show()
```
